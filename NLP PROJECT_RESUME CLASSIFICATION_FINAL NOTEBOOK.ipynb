## 1. React JS

```python
import re
def remove_url(text):
    return re.sub('((www.[^s]+)|(https?://[^s]+))|(http?://[^s]+)', '',text)
def remove_bullet_marks(text):
    bullet_marks = ['-', '*', '●', '‣','.',',',':',"&","'",'"','(',')','-','.']  # Add more bullet marks as needed
    for bullet in bullet_marks:
        text = text.replace(bullet, '')
    return text
```

## Basic EDA

```python
js['Resumes'] = js['Resumess'].apply(remove_url)
```

```python
js['Resumes'] = js['Resumes'].apply(remove_bullet_marks)
```

```python
js['Resumes']=js['Resumes'].apply(lambda x: x.lower())
```

```python
from nltk.corpus import stopwords
stopwords_set = set(stopwords.words('english'))
def cleaning_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in stopwords_set])
```

```python
js['Resumes'] = js['Resumes'].apply(cleaning_stopwords)
```

# Model Building

## 1.Naive Bayes Classifier (using regularization Parameter)

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report


# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(Rws['Resumes'], Rws['Category'], test_size=0.3, random_state=42)

# Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the vectorizer on the training data
X_train_vec = vectorizer.fit_transform(X_train)

# Transform the testing data
X_test_vec = vectorizer.transform(X_test)

# Create and train the Naive Bayes classifier with regularization
nb_classifier = MultinomialNB(alpha=6)  # Adjust the alpha value for regularization
nb_classifier.fit(X_train_vec, y_train)

# Predict the categories for both training and testing data
train_predictions = nb_classifier.predict(X_train_vec)
test_predictions = nb_classifier.predict(X_test_vec)

# Calculate and print the train and test accuracy
train_accuracy_nv = accuracy_score(y_train, train_predictions)
test_accuracy_nv = accuracy_score(y_test, test_predictions)
print("Train Accuracy:", train_accuracy_nv)
print("Test Accuracy:", test_accuracy_nv)

# Generate the classification report for the testing data
report = classification_report(y_test, test_predictions)
print("Classification Report:\n", report)
```

Overall, the model shows good performance for some categories (e.g., "Peoplesoft","React JS" and "Workday") with high precision, recall, and F1-scores. However, it struggles with the "SQL Dev." category, indicating that further improvement might be needed for this specific category.

# Checking Accuracy through user input using Random Forest

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Fit the vectorizer on the entire dataset
X_vec = vectorizer.fit_transform(Rws['Resumes'])
y = Rws['Category']

# Create and train the Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100)  # Adjust the parameters as needed
rf_classifier.fit(X_vec, y)

# User input for a resume
user_input = input("Enter a resume: ")

# Vectorize the user input
user_input_vec = vectorizer.transform([user_input])

# Predict the probabilities for each category
category_probs = rf_classifier.predict_proba(user_input_vec)


# Sample multiple random indices and calculate accuracy for each sampled prediction
num_samples = 100  # Number of samples to generate
accuracies = []
predictions = []

for _ in range(num_samples):
    # Add some randomness to the prediction
    random_index = np.random.choice(range(len(category_probs[0])), p=category_probs[0])
    prediction = rf_classifier.classes_[random_index]

    # Calculate accuracy for the sampled prediction
    user_input_accuracy = rf_classifier.score(user_input_vec, [prediction])
    accuracies.append(user_input_accuracy)
    predictions.append(prediction)

# Calculate the average accuracy
average_accuracy = np.mean(accuracies)

# Print the predicted category and average accuracy
print("Predicted Category:", predictions[0])
print("Average User Input Accuracy:", average_accuracy)
```
